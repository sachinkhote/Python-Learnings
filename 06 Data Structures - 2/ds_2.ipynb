### Joins in sets

```python
# joining two sets using UNION
setA = {1, 2, 3}
setB = {3, 4, 5}
setC = {5, 6, 7}
union_set = setA | setB             # using | operator for union
union_set1 = setB.union(setC)       # using union() method for union
print(union_set)   # {1, 2, 3, 4, 5}
print(union_set1)  # {3, 4, 5, 6, 7}
```

```python
# joining using INTERSECTION
inter_set = setA & setB              # using & operator for intersection
inter_set2 = setB.intersection(setC)    # using intersection() method for intersection
print(inter_set)    # {3}
print(inter_set2)   # {5}
```

```python
# Difference of sets
diff_set = setA - setB               # using - operator for difference
diff_set2 = setB.difference(setC)    # using difference() method for difference

print(diff_set)    # {1, 2}
print(diff_set2)   # {3, 4}
```

```python
# symmertric difference
sym_diff = setA ^ setB                          # using ^ operator for symmetric difference
sym_diff2 = setB.symmetric_difference(setC)     # using symmetric_difference() method for symmetric difference
print(sym_diff)    # {1, 2, 4, 5}
print(sym_diff2)   # {3, 4, 6, 7}
```

### setdefault()
#### used for setting up any default value for desired key

```python
my_dict = {'stud1':{'name': 'sachin', 'class': '10th' },
            'stud2':{'name': 'rahul', 'class': '10th','age':15}}
my_dict['stud1'].setdefault('age', 16)  # set default age for stud1
my_dict['stud2'].setdefault('age', 16)  # set default age for stud2 (already has age, so unchanged)
print(my_dict)

# in this case we have to manually set the dafault for each student 
```

```python
# using loop to set default age for all students
my_dict = {'stud1':{'name': 'sachin', 'class': '10th' }, 
            'stud2':{'name': 'rahul', 'class': '10th','age':15},
            'stud3':{'name': 'virat', 'class': '9th'},
            'stud4':{'name': 'dhoni', 'class': '11th'}}

for student in my_dict:
    my_dict[student].setdefault('age', 16)  # set default age 16 for all students who don't have age

print(my_dict)
```

```python
my_dict = {'name':'sachin', 'class': '10th', }
my_dict.setdefault('age')           # if 'age' key exists, returns its value, else sets it to None
print(my_dict)
```

```python
# Merging two dictionaries representing sales data from two stores
store1_sales = {"Laptop": 15, "Smartphone": 30, "Tablet": 10, "Headphones": 20}
store2_sales = {"Smartphone": 50, "Tablet": 40, "Camera": 25, "Headphones": 30}

# Write your code here
for product, sales in store2_sales.items():
    if product in store1_sales:
        store1_sales[product] +=sales
    else:
        store1_sales[product] =sales
print("Merged Sales Data:",store1_sales)
```

```python
print("Max Product:", max(store1_sales, key=store1_sales.get),'with Total Sales:',max(store1_sales.values()))
```

```python
# create a sorted list that contains each key repeated according to its frequency.
freq_dict = {33: 2, 91: 4, 5: 1, 20: 3, 4: 0}
result= []
for k1, v1 in freq_dict.items():
    result.extend([k1] * v1)
    result.sort()
print(result)
```
