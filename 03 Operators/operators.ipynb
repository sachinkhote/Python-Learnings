```python
# comparison operators
x= 5
y=10

print(x < y)  # True
print(x <= y) # True
print(x > y)  # False
print(x >= y) # False
print(x == y) # False
print(x != y) # True
```

```python
# logical operators
a = True
b = False
print(a and b) # False
print(a or b)  # True
print(not a)   # False
print(not b)   # True
```

```python
#Assignment operators
x = 5
x += 3  # x = x + 3
print(x)  # 8
x *= 2  # x = x * 2
print(x)  # 16
x -= 4  # x = x - 4
print(x)  # 12
x /= 3  # x = x / 3
print(x)  # 4.0
x //= 2  # x = x // 2
print(x)  # 2.0
x %= 2  # x = x % 2
print(x)  # 0.0
```

```python
#membership operators
my_list = [1, 2, 3, 4, 5]
print(3 in my_list)  # True
print(6 in my_list)  # False
print(6 not in my_list)  # True
```

```python
#identity operators
var_1 = 12
var_2 = 12
print(var_1 is var_2)  # True
print(var_1 is not var_2)  # False
```
