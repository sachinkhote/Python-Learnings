### Upper, Lower, Title case and Capitalize

```python
name_upper=var.upper()
print(name_upper)       # converting to upper case

name_lower=var.lower()
print(name_lower)       # converting to lower case

name_title=var.title()
print(name_title)       # converting to title case

name_cap=var.capitalize()
print(name_cap)     # converting to capitalized case
```

### Index and Find

```python
x= "hello sachin bhai, sachin is a good boy"
```
