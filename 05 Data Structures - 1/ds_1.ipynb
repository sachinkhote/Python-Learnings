### add list items

```python
# 1. append()
names.append('pant')    # adding item at last
print(names)

# 2. insert()
names.insert(2, 'rahul')  # adding item at specific position    
print(names)

# 3. extend()
names2 = ['hardik', 'ravindra']
names.extend(names2)            # adding multiple items at last by another list
names.extend(['ishant', 'shami'])  # adding multiple items at last
print(names)

```

```python
# comaprison
l1= [1, 2, 3]
l2 =['1', '2', '3']

print(l1==l2)   # False 
print(l1!=l2)  # True
# print(l1<l2) raise error coz string and int cannot be compared
```

```python
# is operator
ll1, ll2= [1, 2, 3] , [1, 2, 3]
print(ll1 is ll2)  # False, coz both have different memory locations

a1 = 5
a2 = 5
print(a1 is a2)   # True, coz both have same memory locations

# so whenever a list is created, a new memory location is assigned to it even if the contents are same
```

### Remove List items

```python
#1 remove() - removes first matching item
names.remove('rahul')
print(names)

#2 pop() - removes item at specific index
names.pop(3)
print(names)

#3 clear() - removes all items ,but empty list is present
l2.clear()
print(l2)

#4 del 
```

```python
# joining lists using join()
words = ['Hello', 'world', 'from', 'Python']
sentence = ' '.join(words)
print(sentence) 

sentence2 = '-'.join(words)
print(sentence2)
```

```python
# counting occurrences of an item
nums = [1, 2, 3, 1, 4, 1, 5, 2]
count_1 = nums.count(1)
print(f"Number 1 occurs {count_1} times in the list.")
```

```python
# copying List 
#1 using copy() method
original_list = [1, 2, 3, 4, 5]
copied_list = original_list.copy()
print("Original List:", original_list)
print("Copied List:", copied_list)

#2 using slicing
sliced_copied_list = original_list[:]
print("Sliced Copied List:", sliced_copied_list)

#3 using list() constructor
constructor_copied_list = list(original_list)
print("Constructor Copied List:", constructor_copied_list)
```

### Length, Min, Max, Sum

```python
tp = (1 , 2 ,(3,4),5 , 6)
length = len(tp)
print("Length = ",length)

tp2 = (10, 20, 30, 40)
Mini = min(tp2)
print("Minimum = ",Mini)        #min max can only be used integers or floats

maxi = max(tp2)
print("Maximum = ",maxi)

sum1 = sum(tp2)
print("Sum = ",sum1)
```

```python
# chnaging tuple items
tp3 = ('apple', 'banana', 'cherry')
y = list(tp3)
y.append('orange')
tp3 = tuple(y)
print(tp3)

# we convert tuple to list, change the list and convert it back to tuple
```

```python
# adding tuples
tp4 = ('kiwi', 'guava')
tp3 += tp4
print(tp3) 
```

```python
# tuples are immutable , but list within tuple can be changed
tp5 = (1, 2, [3, 4], 5)
tp5[2][0] = 30
print(tp5)
```

```python
# tuple unpacking 
my_tp = (1 ,5, 6)
a,b,c= my_tp
print(a)
print(b)
print(c)
```

```python
my_tp = (1 ,5, 6, 7 ,9, 4)
a, *b, c = my_tp
print(a)    # prints first item
print(b)    # prints middle items as list
print(c)    # prints last item
```

```python
# take two 3d coordinates and perform dot product
coord1 = list(map(int, input().split(' ')))     # read space separated integers
coord2 = list(map(int, input().split()))        # by default split() splits input by space    
print(coord1)
print(coord2)
dot_prod = coord1[0]*coord2[0] + coord1[1]*coord2[1] + coord1[2]*coord2[2]
print("Dot Product:", dot_prod)
```
